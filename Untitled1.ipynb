```python
import cv2
import numpy as np
import face_recognition
import os
import serial
import time
from datetime import datetime
from IPython.display import display, clear_output, Image

# Serial connection to ESP32 (update with the actual port)
esp32_serial = serial.Serial('COM5', 115200)  # Replace 'COM5' with your actual serial port

# Initialize external camera (try different indices if needed)
video_capture = cv2.VideoCapture(1)  # Change to the correct index for your external camera

# Path to training images
path = 'Training_images_folder'
classNames = []
encodeListKnown = []

# Load training images and encode
for person_name in os.listdir(path):
    person_folder = os.path.join(path, person_name)
    if os.path.isdir(person_folder):
        encodings = []
        for img_name in os.listdir(person_folder):
            img_path = os.path.join(person_folder, img_name)
            img = cv2.imread(img_path)
            if img is not None:
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                face_enc = face_recognition.face_encodings(img)
                if face_enc:
                    encodings.append(face_enc[0])
        if encodings:
            avg_encoding = np.mean(encodings, axis=0)
            encodeListKnown.append(avg_encoding)
            classNames.append(person_name)

if not encodeListKnown:
    raise ValueError("No valid face encodings found. Please check the training images.")
print(f"Encoding complete for {len(classNames)} classes.")

# Variables
lock_status = False
recognition_interval = 7
last_recognition_time = 0
recognition_threshold = 0.4

def check_lock_status():
    global lock_status
    if esp32_serial.in_waiting > 0:
        lock_command = esp32_serial.read().decode('utf-8')
        if lock_command == '1':
            lock_status = False
            print("Door unlocked.")
        elif lock_command == '0':
            lock_status = True
            print("Door locked.")

def markAttendance(name):
    if not os.path.exists('Attendance.csv'):
        with open('Attendance.csv', 'w') as f:
            f.writelines('Name,Time\n')
    
    with open('Attendance.csv', 'r+') as f:
        myDataList = f.readlines()
        nameList = [line.split(',')[0] for line in myDataList]
        if name not in nameList:
            now = datetime.now()
            dtString = now.strftime('%H:%M:%S')
            f.writelines(f'{name},{dtString}\n')

def handle_face_recognition(frame):
    global last_recognition_time

    current_time = time.time()
    if current_time - last_recognition_time < recognition_interval:
        return frame

    imgS = cv2.resize(frame, (0, 0), None, 0.25, 0.25)
    imgS = cv2.cvtColor(imgS, cv2.COLOR_BGR2RGB)

    facesCurFrame = face_recognition.face_locations(imgS)
    encodesCurFrame = face_recognition.face_encodings(imgS, facesCurFrame)

    recognized = False

    for encodeFace, faceLoc in zip(encodesCurFrame, facesCurFrame):
        faceDis = face_recognition.face_distance(encodeListKnown, encodeFace)
        minDis = np.min(faceDis)

        top, right, bottom, left = [v * 4 for v in faceLoc]

        if minDis < recognition_threshold:
            matchIndex = np.argmin(faceDis)
            name = classNames[matchIndex]
            recognized = True

            cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
            cv2.putText(frame, name, (left + 6, bottom - 6), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

            markAttendance(name)
            esp32_serial.write(b'G')
            esp32_serial.write(b'O')
            now = datetime.now()
            time_str = now.strftime('%H:%M:%S')
            print(f"[INFO] {name} entered at {time_str}.")
            print(f"[ACTION] Door opening for {name}...")

            time.sleep(5)
            esp32_serial.write(b'C')
            print("[ACTION] Door closing...")
            esp32_serial.write(b'0')
            last_recognition_time = current_time
        else:
            continue

    if not recognized:
        for faceLoc in facesCurFrame:
            top, right, bottom, left = [v * 4 for v in faceLoc]
            cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)
            cv2.putText(frame, "Unknown", (left + 6, bottom - 6), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)
        esp32_serial.write(b'R')
        print("[WARNING] Unrecognized face detected.")

    return frame

def show_frame_in_notebook(frame):
    overlay = frame.copy()
    cv2.rectangle(overlay, (0, 0), (frame.shape[1], 60), (0, 0, 0), -1)
    alpha = 0.6
    cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)

    text = "Face Recognition Smart Lock System"
    status = "Locked" if lock_status else "Unlocked"
    cv2.putText(frame, text, (10, 35), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
    cv2.putText(frame, f"Status: {status}", (frame.shape[1] - 200, 35), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)

    _, img_encoded = cv2.imencode('.jpg', frame)
    clear_output(wait=True)
    display(Image(data=img_encoded.tobytes()))

def main():
    try:
        while True:
            ret, frame = video_capture.read()
            if not ret:
                print("Failed to capture frame.")
                break

            check_lock_status()

            if not lock_status:
                frame = handle_face_recognition(frame)
            else:
                print("[INFO] Door is locked. Face recognition disabled.")

            show_frame_in_notebook(frame)

    except KeyboardInterrupt:
        print("[INFO] Program interrupted by user.")
    
    finally:
        video_capture.release()
        esp32_serial.close()
        print("[INFO] Resources released.")

if __name__ == "__main__":
    main()
```

```python
import cv2
import numpy as np
import face_recognition
import os
import serial
import time
from datetime import datetime
from IPython.display import display, clear_output, Image

# Serial connection to ESP32 (update with the actual port)
esp32_serial = serial.Serial('COM5', 115200)  # Replace 'COM5' with your actual serial port

# Initialize external camera
video_capture = cv2.VideoCapture(1)  # Adjust the index based on your external camera

# Path to training images
path = 'Training_images_folder'
classNames = []
encodeListKnown = []

# Load training images and encode
for person_name in os.listdir(path):
    person_folder = os.path.join(path, person_name)
    if os.path.isdir(person_folder):
        encodings = []
        for img_name in os.listdir(person_folder):
            img_path = os.path.join(person_folder, img_name)
            img = cv2.imread(img_path)
            if img is not None:
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                face_enc = face_recognition.face_encodings(img)
                if face_enc:
                    encodings.append(face_enc[0])
        if encodings:
            avg_encoding = np.mean(encodings, axis=0)
            encodeListKnown.append(avg_encoding)
            classNames.append(person_name)

if not encodeListKnown:
    raise ValueError("No valid face encodings found. Please check the training images.")
print(f"Encoding complete for {len(classNames)} classes.")

# Variables
lock_status = False
recognition_interval = 5  # in seconds
last_recognition_time = 0
recognition_threshold = 0.4

def check_lock_status():
    global lock_status
    if esp32_serial.in_waiting > 0:
        lock_command = esp32_serial.read().decode('utf-8')
        if lock_command == '1':
            lock_status = False
            print("Door unlocked.")
        elif lock_command == '0':
            lock_status = True
            print("Door locked.")

def markAttendance(name):
    if not os.path.exists('Attendance.csv'):
        with open('Attendance.csv', 'w') as f:
            f.writelines('Name,Time\n')
    
    with open('Attendance.csv', 'r+') as f:
        myDataList = f.readlines()
        nameList = [line.split(',')[0] for line in myDataList]
        if name not in nameList:
            now = datetime.now()
            dtString = now.strftime('%H:%M:%S')
            f.writelines(f'{name},{dtString}\n')

def freeze_frame(frame, text, duration=5):
    """
    Display a frame with a message overlayed for a given duration.
    """
    overlay = frame.copy()
    cv2.rectangle(overlay, (0, 0), (frame.shape[1], 60), (0, 0, 0), -1)
    cv2.addWeighted(overlay, 0.6, frame, 0.4, 0, frame)
    cv2.putText(frame, text, (50, frame.shape[0] // 2), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 4, cv2.LINE_AA)

    for _ in range(int(duration * 10)):  # Freeze the frame for the specified duration
        _, img_encoded = cv2.imencode('.jpg', frame)
        clear_output(wait=True)
        display(Image(data=img_encoded.tobytes()))
        time.sleep(0.1)

def handle_face_recognition(frame):
    global last_recognition_time

    current_time = time.time()
    if current_time - last_recognition_time < recognition_interval:
        return frame

    imgS = cv2.resize(frame, (0, 0), None, 0.25, 0.25)
    imgS = cv2.cvtColor(imgS, cv2.COLOR_BGR2RGB)

    facesCurFrame = face_recognition.face_locations(imgS)
    encodesCurFrame = face_recognition.face_encodings(imgS, facesCurFrame)

    for encodeFace, faceLoc in zip(encodesCurFrame, facesCurFrame):
        faceDis = face_recognition.face_distance(encodeListKnown, encodeFace)
        minDis = np.min(faceDis)

        top, right, bottom, left = [v * 4 for v in faceLoc]

        if minDis < recognition_threshold:
            matchIndex = np.argmin(faceDis)
            recognized_name = classNames[matchIndex]

            # Mark attendance and send command to ESP32
            markAttendance(recognized_name)
            esp32_serial.write(b'G')  # Signal for recognized face
            esp32_serial.write(b'O')  # Door open command
            now = datetime.now()
            time_str = now.strftime('%H:%M:%S')
            print(f"[INFO] {recognized_name} entered at {time_str}.")
            print(f"[ACTION] Door opening for {recognized_name}...")

            # Freeze the frame and display recognition message
            freeze_frame(frame, f"{recognized_name} recognized\nEntering...", 5)

            # Pause and close the door
            time.sleep(7)
            esp32_serial.write(b'C')  # Door close command
            print("[ACTION] Door closing...")
            esp32_serial.write(b'0')  # Lock command

            # Update recognition timing
            last_recognition_time = current_time
            break
        else:
            # Display "Unknown Person" message for unrecognized faces
            cv2.rectangle(frame, (left, top), (right, bottom), (0, 0, 255), 2)
            cv2.putText(frame, "Unknown Person", (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (0, 0, 255), 2)

    return frame


def show_frame_in_notebook(frame):
    overlay = frame.copy()
    cv2.rectangle(overlay, (0, 0), (frame.shape[1], 60), (0, 0, 0), -1)
    alpha = 0.6
    cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)

    text = "Smart Door Lock System"
    status = "Locked" if lock_status else "Unlocked"
    cv2.putText(frame, text, (10, 35), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
    cv2.putText(frame, f"Status: {status}", (frame.shape[1] - 200, 35), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)

    _, img_encoded = cv2.imencode('.jpg', frame)
    clear_output(wait=True)
    display(Image(data=img_encoded.tobytes()))

def main():
    try:
        while True:
            ret, frame = video_capture.read()
            if not ret:
                print("Failed to capture frame.")
                break

            check_lock_status()

            if not lock_status:
                frame = handle_face_recognition(frame)
            else:
                print("[INFO] Door is locked. Face recognition disabled.")

            show_frame_in_notebook(frame)

    except KeyboardInterrupt:
        print("[INFO] Program interrupted by user.")
    
    finally:
        video_capture.release()
        esp32_serial.close()
        print("[INFO] Resources released.")

if __name__ == "__main__":
    main()
```
